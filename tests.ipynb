```python
import pylab as pl
import pandas as pd
import numpy as np
import scipy.stats
%pylab inline
```

```python
x = np.arange(0.,100.) - 50

mu = 10.
stdev = 3.
pl.plot(x, scipy.stats.norm.pdf(x, loc=mu, scale=stdev))
pl.plot(x, 1./stdev/np.sqrt(2.*np.pi)*np.exp(-0.5* (x-mu)*(x-mu)/stdev/stdev),'r--')
```

```python
dist_n = np.random.randn(1000)
print scipy.stats.kstest(dist_n, 'norm')
print scipy.stats.anderson(dist_n, dist='norm')
hs, bs, ps = pl.hist(dist_n)
print hs, bs, ps
pl.plot(bs[:-1]+0.5*(bs[1]-bs[0]), hs)
```

```python
pl.plot(bs[1:], hs.cumsum()/max(hs.cumsum()), 'b')
print scipy.stats.norm.cdf(bs)
pl.plot(bs[1:]-0.5*(bs[1]-bs[0]),scipy.stats.norm.cdf(bs[:-1]+0.5*(bs[1]-bs[0])), 'r--')
```
